```python
import seaborn as sns

# Seaborn 스타일 적용
sns.set_style("whitegrid")
uploaded_image_path ='./result/before_inference/6e9a310d-9481-4b65-996d-cb8bc5aaca1c.jpg'
image = Image.open(uploaded_image_path)
img_width, img_height = image.size

# 저장할 경로 설정
output_path_seaborn =  "./results_seaborn3.jpg"


label_colors = sns.color_palette("hsv", n_colors=11)

# Map disease names to specific colors
disease_color_map = {
    "Caries": label_colors[0],
    "Gingivitis": label_colors[1],
    "Periodontitis": label_colors[2],
    "Oral Cancer": label_colors[3],
    "Erosion": label_colors[4],
    "Tartar": label_colors[5],
    "Fracture": label_colors[6],
    "Ulcer": label_colors[7],
    "Abscess": label_colors[8],
    "Discoloration": label_colors[9],
    "Other": label_colors[10]
}

# 시각화 및 저장 (업데이트된 스타일링)
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(image)

# YOLO 데이터를 바탕으로 박스 그리기
for tooth, diseases in data['tooth_diseases'].items():
    for disease in diseases:
        x_min, y_min, box_width, box_height = yolo_to_pixel(disease['location'], img_width, img_height)
        confidence = disease['confidence']
        disease_name = disease['disease_name']

        # 박스 및 텍스트 색상 설정
        color = disease_color_map.get(disease_name, "lime")

        # 박스 추가
        rect = patches.Rectangle(
            (x_min, y_min), box_width, box_height, linewidth=2, edgecolor=color, facecolor='none', alpha=0.9
        )
        ax.add_patch(rect)

        # 텍스트 추가 (위쪽에 배치)
        ax.text(
            x_min, y_min - 5, f"{disease_name} ({confidence:.2f})",
            fontsize=12, color='white', ha='left', va='bottom',
            bbox=dict(facecolor=color, alpha=0.8, edgecolor='none', boxstyle='round,pad=0.3')
        )

# 스타일 설정
ax.axis('off')

# 이미지 저장
plt.savefig(output_path_seaborn, bbox_inches='tight', pad_inches=0, dpi=300)
plt.close(fig)

print(f"Final detection results with label-specific colors saved at {output_path_seaborn}")
```
